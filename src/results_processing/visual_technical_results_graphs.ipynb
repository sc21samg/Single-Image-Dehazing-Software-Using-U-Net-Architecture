```python
from PIL import Image, ImageDraw, ImageFont

# Mount Google Drive if necessary
from google.colab import drive
drive.mount("/content/drive", force_remount=True)

# Define image paths for the first and second row of images
image_paths = [
    "/content/drive/My Drive/diss/myproj/data_test/test/hamburg_000000_000042_leftImg8bit_foggy_beta_0.005.png",
    "/content/drive/My Drive/diss/myproj/data_test/test/hamburg_000000_000042_leftImg8bit_foggy_beta_0.01.png",
    "/content/drive/My Drive/diss/myproj/data_test/test/hamburg_000000_000042_leftImg8bit_foggy_beta_0.02.png",
]

save_paths = [
    "/content/drive/My Drive/diss/myproj/results/images/hamburg_000000_000042_leftImg8bit_foggy_beta_0.005.png",
    "/content/drive/My Drive/diss/myproj/results/images/hamburg_000000_000042_leftImg8bit_foggy_beta_0.01.png",
    "/content/drive/My Drive/diss/myproj/results/images/hamburg_000000_000042_leftImg8bit_foggy_beta_0.02.png",
]

# Define labels for each image
labels = ["Beta 0.005", "Beta 0.01", "Beta 0.02"]

# Load all images for the first and second rows at target size (256, 256)
images1 = [Image.open(path).resize((256, 256)) for path in image_paths]
images2 = [Image.open(path).resize((256, 256)) for path in save_paths]

# Calculate the total width (maximum of both rows) and total height (sum of both rows' heights)
total_width = max(sum(img.width for img in images1), sum(img.width for img in images2))
max_height1 = max(img.height for img in images1)
max_height2 = max(img.height for img in images2)

# Assume font size of 80 and calculate the approximate height required for the text, adding some padding
font_size = 50
text_padding = 10  # Reduced padding

# Create a new blank image with additional height to accommodate text and two rows of images
new_image = Image.new('RGB', (total_width, max_height1 + max_height2 + font_size + text_padding * 2), "white")

# Paste images into new image in two rows
current_width = 0
for image in images1:
    new_image.paste(image, (current_width, 0))
    current_width += image.width

current_width = 0
for image in images2:
    new_image.paste(image, (current_width, max_height1 + text_padding))  # Reduced vertical space between rows
    current_width += image.width

# Initialize ImageDraw
draw = ImageDraw.Draw(new_image)

# Load font
font_path = '/usr/share/fonts/truetype/liberation/LiberationSans-Regular.ttf'
try:
    font = ImageFont.truetype(font_path, font_size)
except IOError:
    font = ImageFont.load_default()

# Define text color
text_color = (0, 0, 0)  # Black color

# Draw text under each image for the second row only
current_width = 0
for i, image in enumerate(images2):
    text_width, _ = draw.textsize(labels[i], font=font)
    text_x_position = current_width + (image.width - text_width) / 2
    text_y_position = max_height1 + max_height2   # Adjusted y-position for labels
    draw.text((text_x_position, text_y_position), labels[i], font=font, fill=text_color)
    current_width += image.width

# Save the new image
output_path = "/content/drive/My Drive/diss/myproj/results/stitched_image_with_text.jpg"
new_image.save(output_path)
output_path
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the data from the file
def read_data(file_path):
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            if 'Image:' in line:
                parts = line.split(',')
                image_name = parts[0].split(': ')[1].strip().replace("000000_000042_leftImg8bit_foggy_", "")
                psnr = float(parts[1].split(': ')[1].strip())
                ssim = float(parts[2].split(': ')[1].strip())
                data.append([image_name, psnr, ssim])
    return data

# Create a DataFrame from the data
def create_dataframe(data):
    df = pd.DataFrame(data, columns=['Image Name', 'PSNR', 'SSIM'])
    return df

# Plot the table using Matplotlib with highlighted cells
def plot_table(df):
    fig, ax = plt.subplots(figsize=(12, 2))  # Adjust the figure size as needed
    ax.axis('tight')
    ax.axis('off')
    table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')

    # Finding the max and min values for PSNR and SSIM
    max_psnr = df['PSNR'].max()
    min_psnr = df['PSNR'].min()
    max_ssim = df['SSIM'].max()
    min_ssim = df['SSIM'].min()

    # Color cells based on the highest and lowest values
    for i in range(len(df)):
        for j, col in enumerate(df.columns):
            cell = table[(i+1, j)]
            value = df.iloc[i][col]
            if col == 'PSNR':
                if value == max_psnr:
                    cell.set_facecolor('blue')
                    cell.set_text_props(color='white', weight='bold')
                elif value == min_psnr:
                    cell.set_facecolor('red')
                    cell.set_text_props(color='white', weight='bold')
            elif col == 'SSIM':
                if value == max_ssim:
                    cell.set_facecolor('blue')
                    cell.set_text_props(color='white', weight='bold')
                elif value == min_ssim:
                    cell.set_facecolor('red')
                    cell.set_text_props(color='white', weight='bold')

    table.auto_set_font_size(False)
    table.set_fontsize(14)
    table.scale(1.2, 1.2)  # Scale table size to make it more readable

    # Add a legend on the right side of the table
    plt.figtext(1.0, 0.7, 'Legend:', fontsize=12, verticalalignment='top')
    plt.figtext(1.0, 0.55, 'Highest Value', fontsize=14, verticalalignment='top', color='white', weight='bold', bbox=dict(facecolor='blue', edgecolor='black', boxstyle='round,pad=0.5'))
    plt.figtext(1.0, 0.35, 'Lowest Value', fontsize=14, verticalalignment='top', color='white', weight='bold', bbox=dict(facecolor='red', edgecolor='black', boxstyle='round,pad=0.5'))


    # Add title
    plt.title("Images Performance Metrics", fontsize=14, weight='bold')

    # Save the figure
    plt.savefig('/content/drive/My Drive/diss/myproj/results/model_metrics_table.png', bbox_inches='tight', pad_inches=0.05)
    plt.show()

# Main execution
file_path = '/content/drive/My Drive/diss/myproj/results/dehazing_results.txt'
data = read_data(file_path)
df = create_dataframe(data)
plot_table(df)
```
